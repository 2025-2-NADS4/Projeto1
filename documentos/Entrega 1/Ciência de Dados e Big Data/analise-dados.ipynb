```python
import pandas as pd # importa a biblioteca pandas

# Carregar os arquivos CSV.

# lê os arquivos e armazena em uma variavel - dataframe
df_order = pd.read_csv('/content/Order_semicolon.csv', sep=';')
df_campaign_queue = pd.read_csv('/content/CampaignQueue_semicolon.csv', sep=';')
df_campaign = pd.read_csv('/content/Campaign_semicolon.csv', sep=';')

print("Conseguimos carregar os dados com sucesso!")
```

```python
print("df order")
print("------- 5 primeiras linhas -------")
display(df_order.head()) # mostra as 5 primeiras linhas da tabela

print("------- Informações da tabela -------")
df_order.info() # mostra informações da tabela
```

```python
print("df_campaign_queue")
print("------- 5 primeiras linhas -------")
display(df_campaign_queue.head()) # mostra as 5 primeiras linhas da tabela

print("------- Informações da tabela -------")
df_campaign_queue.info() # mostra informações da tabela
```

```python
print("df_campaign")
print("------- 5 primeiras linhas -------")
display(df_campaign.head()) # mostra as 5 primeiras linhas da tabela

print("------- Informações da tabela -------")
df_campaign.info() # mostra informações da tabela
```

```python
# lista de colunas que são de datas para a tabela ORDER
date_cols_order = ['createdAt', 'updatedAt', 'scheduledAt']
# faz a conversão das colunas para o tipo data
df_order[date_cols_order] = df_order[date_cols_order].apply(pd.to_datetime, format='%d/%m/%Y %H:%M', errors='coerce')


# lista de colunas que são de datas para a tabela CAMPAIGN QUEUE
date_cols_queue = ['scheduledAt', 'sendAt', 'createdAt', 'updatedAt']
# faz a conversão das colunas para o tipo data
df_campaign_queue[date_cols_queue] = df_campaign_queue[date_cols_queue].apply(pd.to_datetime, format='%d/%m/%Y %H:%M', errors='coerce')


# lista de colunas que são de datas para a tabela CAMPAIGN
date_cols_campaign = ['createdAt', 'updatedAt']
# faz a conversão das colunas para o tipo data
df_campaign[date_cols_campaign] = df_campaign[date_cols_campaign].apply(pd.to_datetime, format='%d/%m/%Y %H:%M', errors='coerce')


print("\nConversão de datas concluída com sucesso!")
print("---------------------------------------------")
print("\n--- Verificando df_order ---")
df_order.info()

print("\n--- Verificando df_campaign_queue ---")
df_campaign_queue.info()

print("\n--- Verificando df_campaign ---")
df_campaign.info()
```

```python
# importando bibliotecas para visualizar graficos e tabelas
import matplotlib.pyplot as plt
import seaborn as sns

# pegamos a coluna "totalAmount" e ai obtemos um resumo rapido - média, mediana, minimo e maximo
print("------- Resumo do totalAmount -------")
display(df_order['totalAmount'].describe())

# usamos a coluna salesChannel para poder ver os canais de venda e contar os pedidos deles
print("\nContagem de pedidos por canal de venda")
contagem_canal = df_order['salesChannel'].value_counts()
display(contagem_canal)

# grafico de barras para visualizar a contagem por canal
plt.figure(figsize=(10, 6))
# criando graficos de barra horizontal para melhor visualização
sns.countplot(y=df_order['salesChannel'], order=contagem_canal.index, palette='viridis')
plt.title('Quantidade de Pedidos por Canal de Venda')
plt.xlabel('Quantidade de Pedidos')
plt.ylabel('Canal de Venda')
plt.show()

# aqui pegamos a parte do tipo de pedido e fazemos a contagem dos pedidos para cada tipo
print("\nContagem de tipos de pedidos")
contagem_tipo = df_order['orderType'].value_counts()
display(contagem_tipo)

# grafico de barras para visualizar a contagem por tipo
plt.figure(figsize=(8, 5))
sns.countplot(x=df_order['orderType'], order=contagem_tipo.index, palette='plasma')
plt.title('Quantidade de Pedidos por Tipo')
plt.xlabel('Tipo de Pedido')
plt.ylabel('Quantidade de Pedidos')
plt.show()
```

```python
# Analise de Vendas por Canal

print("--- Análise de Faturamento e Ticket Médio por Canal ---")

# Usando groupby pra agrupar os pedidos por canal e o .agg pra calcular tudo de uma vez
analise_por_canal = df_order.groupby('salesChannel')['totalAmount'].agg(['sum', 'count', 'mean'])

# Renomeando as colunas pra tabela ficar mais fácil de ler
analise_por_canal.columns = ['Faturamento Total', 'Qtd de Pedidos', 'Ticket Médio']

# Ordenando pra ver qual canal tem o maior ticket médio
analise_por_canal = analise_por_canal.sort_values(by='Ticket Médio', ascending=False)

# Mostrando a tabela com a análise pronta
display(analise_por_canal)
```

```python
# Análise Vendas por Dia da Semana

print("--- Análise de Vendas por Dia da Semana ---")

# Isso aqui só funciona porque convertemos as datas lá em cima
# Criando uma coluna nova com o nome do dia da semana
df_order['dia_da_semana'] = df_order['createdAt'].dt.day_name()

# Contando quantos pedidos rolaram em cada dia
pedidos_por_dia = df_order['dia_da_semana'].value_counts()

# Deixando na ordem certa (Segunda, Terça, etc.)
dias_ordenados = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
pedidos_por_dia = pedidos_por_dia.reindex(dias_ordenados)

# Agora o gráfico pra visualizar isso
plt.figure(figsize=(12, 7))
sns.barplot(x=pedidos_por_dia.index, y=pedidos_por_dia.values, palette='cubehelix')
plt.title('Quantidade de Pedidos por Dia da Semana')
plt.xlabel('Dia da Semana')
plt.ylabel('Número de Pedidos')
plt.xticks(rotation=45)
plt.show()
```

```python
### Análise Engajamento das Campanhas de Marketing

print("--- Análise de Engajamento das Campanhas ---")

# Juntando (merge) a tabela de envios com a de campanhas pra gente ter os nomes
df_merged = pd.merge(df_campaign_queue, df_campaign, left_on='campaignId', right_on='id', how='left')

# Criando um mapa pra traduzir o número do status para um texto que a gente entende
status_map = {
    1: 'Agendado',
    2: 'Enviado',
    3: 'Recebido',
    4: 'Lido',
    5: 'Deletado',
    6: 'Pendente'
}
df_merged['status_legivel'] = df_merged['status_x'].map(status_map)

# Contando quantos envios existem para cada status
funil_geral = df_merged['status_legivel'].value_counts()

print("\n### Contagem de Mensagens em Cada Status:")
display(funil_geral)

# Calculando a taxa de leitura geral para ter uma métrica de sucesso
# (Lido / Total de Enviados)
if 'Enviado' in funil_geral.index and 'Lido' in funil_geral.index:
    taxa_leitura = (funil_geral['Lido'] / funil_geral['Enviado']) * 100
    print(f"\nTaxa de Leitura Geral (dos enviados): {taxa_leitura:.2f}%")
else:
    print("\nNão foi possível calcular a taxa de leitura (dados de 'Enviado' ou 'Lido' não encontrados).")
```
